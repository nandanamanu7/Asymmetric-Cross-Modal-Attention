# VQA v2.0 Data Exploration

Load and inspect the dataset structure, answer distribution, and sample image-question-answer triplets.

```python
import sys
sys.path.insert(0, '..')
from pathlib import Path
from data.preprocess import load_questions, load_answers, build_answer_vocab, get_top_k_answers
from data import VQADataset

data_dir = Path('../data')
q_path = data_dir / 'questions' / 'val_questions.json'
a_path = data_dir / 'answers' / 'val_annotations.json'
if not a_path.exists():
    a_path = data_dir / 'answers' / 'val_answers.json'

if q_path.exists() and a_path.exists():
    questions = load_questions(str(q_path))
    answers = load_answers(str(a_path))
    print('Number of questions:', len(questions))
    print('Number of annotations:', len(answers))
    top_k = get_top_k_answers(str(a_path), k=100)
    print('Top 10 answers:', top_k[:10])
else:
    print('Place VQA v2 questions and answers under data/questions/ and data/answers/')
    print('See Detailed_Project_Proposal.md for download links.')
```

```python
# Build dataset with small subset and inspect one batch
if q_path.exists() and a_path.exists():
    ds = VQADataset(
        image_dir=str(data_dir / 'images'),
        questions_file=str(q_path),
        answers_file=str(a_path),
        top_k_answers=1000,
        subset_size=100,
    )
    img, tokens, ans_idx = ds[0]
    print('Image shape:', img.shape)
    print('input_ids shape:', tokens['input_ids'].shape)
    print('Answer index:', ans_idx)
    print('Num classes:', ds.num_answers)
```
